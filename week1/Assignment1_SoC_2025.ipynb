```python
from PIL import Image
import os

out_dir = 'out'
resized_dir = 'out_small'
os.makedirs(resized_dir, exist_ok=True)

for fname in os.listdir(out_dir):
    if fname.endswith('.png'):
        img = Image.open(os.path.join(out_dir, fname))
        img = img.resize((256, 256))  # change size here
        img.save(os.path.join(resized_dir, fname))
```

```python
import numpy as np
Z = np.random.randn(512)
```

```python
import torch
device = torch.device('cuda')
network_path = 'models/ffhq.pkl'
```

```python
import legacy
with open(network_path, 'rb') as f:
    G = legacy.load_network_pkl(f)['G_ema'].to(device)
```

```python
z = torch.from_numpy(np.random.randn(10, G.z_dim)).to(device)
```

```python
w_plus = G.mapping(z, None)
```

```python
print(f"Z shape: {z.shape}")
print(f"W+ shape: {w_plus.shape}")
```

```python
images = G.synthesis(w_plus, noise_mode='const')
```

```python
images = (images.clamp(-1, 1) + 1) / 2.0
```

```python
import matplotlib.pyplot as plt
from torchvision.transforms.functional import to_pil_image
for i in range(5):
    img = to_pil_image(images[i].cpu())
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Generated Face {i+1}")
    plt.show()
```

```python
num_samples = 10
```

```python
os.makedirs('latents', exist_ok=True)
os.makedirs('generated_images', exist_ok=True)
```

```python
for i in range(num_samples):
    # Save Z
    np.save(f'latents/z_{i:02d}.npy', z[i].cpu().numpy())

    # Save W+
    np.save(f'latents/w+_{i:02d}.npy', w_plus[i].cpu().numpy())

    # Convert and save image
    img = (images[i].cpu().numpy() * 255).astype(np.uint8)
    img = np.transpose(img, (1, 2, 0))  # CHW → HWC
    Image.fromarray(img).save(f'generated_images/image_{i:02d}.png')
```
